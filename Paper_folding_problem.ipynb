## **[Problem 3] Create using a for statement**
Next, create a program that uses the for statement.

Do not use exponentiation arithmetic operators. Only the four arithmetic operations+ , - , * , / ) are allowed.

```python
number1 = [6, 14, 10]
number2 = [2, 4, 8]

add = []
sub = []
multi = []
div = []

for num in range(3):
    add.append( number1[num] + number2[num])
    sub.append( number1[num] - number2[num])
    multi.append( number1[num] * number2[num])
    div.append( number1[num] / number2[num])

print("\nThe result of Addition =  ", add)
print("The result of Subtraction =  ", sub)
print("The result of Multiplication =  ", multi)
print("The result of Division =  ", div)
```
